# Multiple Linear Regression

```python
# Importing Required Libraries To Train Model
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
# Reading data
df = pd.read_csv('../datasets/Regression/multiple_linear_regression_dataset.csv')
df.head()
```

```python
# Splitting Data For Train And Test
msk = np.random.rand(len(df)) < 0.75
train = df[msk]
test = df[~msk]
```

```python
train_x = np.asanyarray(train[['age','experience']])
train_y = np.asanyarray(train[['income']])
regr = linear_model.LinearRegression()
regr.fit(train_x,train_y)

print("Intercept %.2f" % regr.intercept_[0])
print("coefficient: %.2f" % regr.coef_[0][0])
```

```python
# testing model
test_x = np.asanyarray(test[['age','experience']])
test_y = np.asanyarray(test[['income']])
print("MSE : %.2f" % np.mean((regr.predict(test_x) - test_y)**2))
print("Varience - Score : %.2f" % regr.score(test_x, test_y))
```
